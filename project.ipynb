Q4 How do we clean the gross colum and vote column


```python
df["gross_income"]= df["gross_income"][df["gross_income"].notna()].str.replace(",","", regex=False)
df["gross_income"]= df["gross_income"][df["gross_income"].notna()].str.replace('$',"", regex=False)
df["gross_income"]= df["gross_income"][df["gross_income"].notna()].str.replace('m',"", regex=False)
df["gross_income"]= df["gross_income"][df["gross_income"].notna()].str.replace('M',"", regex=False)
```

```python
df["gross_income"] = df["gross_income"].astype(float)
```

```python
df['votes'] = df['votes'].str.replace(',', '', regex=False)
df['votes'] = df['votes'].astype(int)
```

```python
# Q6: Find hidden gems (high rating, low votes)
# Using 75th percentile for high rating and 25th percentile for low votes based on df.describe()
high_rating_threshold_gem = df['rating'].quantile(0.75)
low_votes_threshold_gem = df['votes'].quantile(0.25)

hidden_gems = df[(df['rating'] > high_rating_threshold_gem) & (df['votes'] < low_votes_threshold_gem)]
print(f"Number of hidden gems found: {len(hidden_gems)}")
hidden_gems.head()
```

Q7 find the overhyped movies (low rating high gross)

```python
# Q7: Find overhyped movies (low rating, high gross income)
# Using 25th percentile for low rating and 75th percentile for high gross income based on df.describe()
low_rating_threshold = df['rating'].quantile(0.25)
high_gross_threshold = df['gross_income'].quantile(0.75)

overhyped_movies = df[(df['rating'] < low_rating_threshold) & (df['gross_income'] > high_gross_threshold)]
print(f"Number of overhyped movies found: {len(overhyped_movies)}")
overhyped_movies.head()
```
